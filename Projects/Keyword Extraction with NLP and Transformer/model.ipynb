```python
### now need to perfrom text preprocessing 

"""
1. lower case
2. remove html
3. remove url
4. remove special char
5. remove stop word
6. remove words less than three letters
7. lemmatization
8. stemmatization
"""
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
stop_words=stopwords.words('english')
```

```python
def text_preprocessing(txt):

    ###1 lower case
    txt=txt.lower()

    ###2 remove html
    txt=re.sub(r'<.*?>',' ',txt)

    ###3 remove special chhar
    txt=re.sub(r'[^a-zA-Z]',' ',txt)

    ###4 toeknization
    txt=nltk.word_tokenize(txt)

    ### 5 remove stopwords

    txt=[word for word in txt if word not in stop_words]

    ###6 if word len less than 3 then remove 

    txt=[word for word in txt if len(word)>3]

    ###7 lemmatization

    stemming=PorterStemmer()
    txt=[stemming.stem(word) for word in txt]
    
    return ' '.join(txt)
```

```python
docs=df['paper_text'].apply(lambda x:text_preprocessing(x))
### text preprocessing done
```

```python
### now do vectorization
### it calucateb tje counts of the words in the docs

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_df=0.95,max_features=5000,ngram_range=(1,3))
word_count_vector=cv.fit_transform(docs)
```

```python
### now use TFIdfTransformer (use for calculate the term frequiecy and score of the word)

from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer=TfidfTransformer(smooth_idf=True,use_idf=True)
tfidf_transformer2=tfidf_transformer.fit(word_count_vector)
```

```python
tfidf_transformer2
```

```python
len(cv.get_feature_names_out())
```

```python
cv.get_feature_names_out()
```

```python
### calcuate the word importance and score
ans = tfidf_transformer2.transform(cv.transform([docs[0]]))
```

```python
# Assuming 'ans' is the result of tfidf_transformer2.transform(cv.transform([docs[0]]))

# Convert the TF-IDF scores to a dense array
tfidf_scores = ans.toarray()[0]

# Get the feature names (words) from the CountVectorizer
feature_names = cv.get_feature_names_out()

# Create a dictionary mapping words to their TF-IDF scores
word_scores = {feature_names[i]: tfidf_scores[i] for i in range(len(feature_names))}

# Sort the dictionary by scores in descending order and get the top 10
top_10_words = sorted(word_scores.items(), key=lambda item: item[1], reverse=True)[:10]

# Print the top 10 words and their scores
for word, score in top_10_words:
    print(f"{word}: {score}")
```
